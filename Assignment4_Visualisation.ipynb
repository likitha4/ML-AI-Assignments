```python
lan=["c","c++","Java","Python","PHP"]
stu=[25,17,34,69,82]
plt.title("NO.OF STUDENTS ENROLLED IN DIFFERENT LANGUAGES")
plt.pie(stu,labels=lan,autopct="%1.2f%%")
plt.show()
```
